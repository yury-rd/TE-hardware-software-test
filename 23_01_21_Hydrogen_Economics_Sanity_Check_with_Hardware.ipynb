# Solar: schedule for a given location

```python
import pvlib
from pvlib import irradiance
from pvlib.location import Location
from pvlib.irradiance import disc
from pvlib.pvsystem import PVSystem
from pvlib.modelchain import ModelChain
from pvlib.temperature import TEMPERATURE_MODEL_PARAMETERS



## Returns normalized per 1KW power
## Based on NREL Computations

class PV_schedule:
  # The init method or constructor
  def __init__(self, location):
    self.pv_azimuth = location['azimuth']
    self.lattitude = location['lattitude']
    self.longitude = location['longitude']
    self.altitude = location['altitude']
    self.surface_tilt = location['tilt']
    self.surface_azimuth = location['azimuth']
    self.timezone = pytz.timezone("US/Pacific")
    self.pv_location = Location(latitude=self.lattitude,
                                longitude=self.longitude,
                                tz=self.timezone, altitude=self.altitude)

    # Yury: Normalization is based on the Hollister annual average (5.2 MWh per MW)
    self.pv_normalization = 9.35e-3

  def get_irradiance(self, datetime_range, cloud_cover=0):
    solar_position = self.pv_location.get_solarposition(datetime_range)

    clearsky = self.pv_location.get_clearsky(datetime_range, model='ineichen')

    # We are using clear sky scaling of GHI based on Larson et al.
    # “Day-ahead forecasting of solar power output from photovoltaic plants in the
    # American Southwest” Renewable Energy 91, 11-20 (2016).
    # From the paper below we take offset = 0.35
    offset = 0.35
    cloud_cover = cloud_cover / 100
    cloudy_ghi = (offset + (1 - offset) * (1 - cloud_cover)) * clearsky['ghi']
    cloudy_dni = disc(cloudy_ghi, solar_position['zenith'], datetime_range)['dni']
    cloudy_dhi = cloudy_ghi - cloudy_dni * np.cos(np.radians(solar_position['zenith']))

    cloudy_POA_irradiance = irradiance.get_total_irradiance(
        surface_tilt=self.surface_tilt,
        surface_azimuth=self.surface_azimuth,
        solar_zenith=solar_position['apparent_zenith'],
        solar_azimuth=solar_position['azimuth'],
        dni=cloudy_dni,
        ghi=cloudy_ghi,
        dhi=cloudy_dhi
    )

    return cloudy_POA_irradiance

  # month_start_datetime = timezone.localize(datetime(2022,7,1,0,0,0))
  def get_pv_power(self, start_datetime, end_datetime):
    full_datetime_range = pd.date_range(start = self.timezone.localize(start_datetime),
                                        end = self.timezone.localize(end_datetime),
                                        closed=None,
                                        tz=self.timezone, freq="5min")

    # use a reasonable default setup
    temperature_model_parameters = TEMPERATURE_MODEL_PARAMETERS['sapm']['open_rack_glass_glass']
    sandia_modules = pvlib.pvsystem.retrieve_sam('SandiaMod')
    sandia_module = sandia_modules['Canadian_Solar_CS5P_220M___2009_']
    cec_inverters = pvlib.pvsystem.retrieve_sam('cecinverter')
    cec_inverter = cec_inverters['ABB__MICRO_0_25_I_OUTD_US_208__208V_']

    pv_location = Location(latitude=self.lattitude, longitude=self.longitude, tz=self.timezone, altitude=self.altitude)

    system = PVSystem(surface_tilt=self.surface_tilt, surface_azimuth=self.surface_azimuth,
                      module_parameters=sandia_module,
                      inverter_parameters=cec_inverter,
                      temperature_model_parameters=temperature_model_parameters)

    mc = ModelChain(system, pv_location)

    solar_irradiance = self.get_irradiance(datetime_range=full_datetime_range)

    weather = pd.DataFrame(data={'poa_global': solar_irradiance["poa_global"],
                           'poa_direct': solar_irradiance["poa_direct"],
                           'poa_diffuse': solar_irradiance["poa_diffuse"],
                           'temp_air': 25,
                           'wind_speed': 0},
                           index=full_datetime_range)

    #display(weather)

    mc.run_model_from_poa(weather)
    df_pv_power = pd.DataFrame(columns=['VALUE'])
    #print("AC", mc.results.ac.values)
    print("AC", mc.results)
    df_pv_power['VALUE'] = mc.results.ac.values*self.pv_normalization
    #df.index = pd.to_datetime(df.index)
    df_pv_power.index = full_datetime_range

    return df_pv_power
```

```python
class DualAxisTrackerMount(pvsystem.AbstractMount):
    def get_orientation(self, solar_zenith, solar_azimuth):
        # no rotation limits, no backtracking
        return {'surface_tilt': solar_zenith, 'surface_azimuth': solar_azimuth}


# schedule generation

def get_pv_schedule(loc, time_zone, start_date, end_date):
  array = pvsystem.Array(
    mount=DualAxisTrackerMount(),
    module_parameters=dict(pdc0=1, gamma_pdc=-0.004, b=0.05),
    temperature_model_parameters=dict(a=-3.56, b=-0.075, deltaT=3))
  system = pvsystem.PVSystem(arrays=[array], inverter_parameters=dict(pdc0=3))
  mc = modelchain.ModelChain(system, loc, spectral_model='no_loss')
  times = pd.date_range(start_date, end_date, freq='5min',
                      tz=time_zone)
  weather = loc.get_clearsky(times)
  mc.run_model(weather)

  ## In case you need to verify
  #mc.results.ac.plot()
  #plt.ylabel('Output Power')
  #plt.show()

  print("AC model", mc.results.ac)

  # account for real loss of PV panels including dust++
  df_ = mc.results.ac*(1 - solar["real_loss"])

  # aggregate
  df = pd.DataFrame(columns=['time', 'value'])
  df['value'] = df_.values
  df['time'] = df_.index
  df['time'] = df['time'].dt.time
  df = df.groupby([df['time']]).value.mean()/12
  # Convert the whole dataframe as a string and display
  #print(df.to_string())
  df.to_csv(root_path+'pv_schedule.csv')

  return df
```

```python
## No Price Escalation over time is considered
# Should be 3% for CA

class Grid:
  def __init__(self):
    self.demand_charge = 0 #per KW
    self.daily_fee = 0 #meter + equipement rental
    self.price = pd.DataFrame([])
    self.montly_max = 1
    self.capex = 0
    self.timezone = pytz.timezone("US/Pacific")
    self.timezone_factor = '-00'

  def get_caiso_price(self, start_date, end_date):
    api_name = 'http://oasis.caiso.com/oasisapi/SingleZip?'
    queryname = 'PRC_INTVL_LMP'
    market_run_id = 'RTM'
    resultformat = '6' # this downloads as csv
    node = 'HOLLISTR_1_N101'
    version = 2

    date = start_date
    while date < end_date:
      dt_start = date.strftime("%Y%m%d") +  'T00:00' + self.timezone_factor +'00'
      date += timedelta(days=1)
      dt_end = date.strftime("%Y%m%d") +  'T00:00' + self.timezone_factor +'00'
      url_single = api_name+'queryname='+queryname+'&'+'startdatetime='+dt_start+'&'+'enddatetime='+dt_end+'&'+'version='+str(version)+'&'+'market_run_id='+market_run_id+'&'+'node='+node+'&'+'resultformat='+resultformat
      r = requests.get(url_single, stream=True, verify=False, timeout=500)
      z = zipfile.ZipFile(io.BytesIO(r.content))
      df = pd.read_csv(z.open(z.namelist()[0]))
      self.price = pd.concat([self.price, df], ignore_index=True)
      sleep(5)

    self.price = self.price.loc[self.price['LMP_TYPE'] == 'LMP']
    #print("self.price")
    #print(self.price)
    self.price.drop(self.price.columns.difference(['INTERVALSTARTTIME_GMT', 'VALUE']), 1, inplace=True)
    self.price = self.price.groupby(['INTERVALSTARTTIME_GMT']).sum()
    self.price = self.price.rename(columns={'INTERVALSTARTTIME_GMT': 'TIME', 'VALUE':'MWh_buy'})
    self.price['MWh_buy'] = self.price['MWh_buy']
    #print(self.price)
    caiso_fee = 1000*(0.0000836 + 0.0002899)*2
    self.price['MWh_buy'] = self.price['MWh_buy'] + caiso_fee
    self.price['MWh_sell'] = self.price['MWh_buy'] - caiso_fee

    #demand charge fee
    self.price['MW_buy'] = 0

    return self.price

  def get_pge_b20_price(self, start_date, end_date):
    i = pd.date_range(start_date, end_date, freq='5min', tz=self.timezone, closed= None)
    ##df[df['A'] > 2]['B'] = new_val
    self.price = pd.DataFrame({'MW_buy': 28040, 'MWh_buy':140, 'MWh_sell':140}, index=i)

    #Peak
    idx = self.price.between_time('16:00', '20:55').index
    self.price.loc[idx, 'MW_buy'] = 28040 + 3130
    self.price.loc[idx, 'MWh_buy'] = 390
    self.price.loc[idx, 'MWh_sell'] = 390

    #Part-Peak 1
    idx = self.price.between_time('14:00', '15:55').index
    self.price.loc[idx, 'MW_buy'] = 28040 + 900
    self.price.loc[idx, 'MWh_buy'] = 200
    self.price.loc[idx, 'MWh_sell'] = 200

    #Part-Peak 2
    idx = self.price.between_time('21:00', '22:55').index
    self.price.loc[idx, 'MW_buy'] = 28040 + 900
    self.price.loc[idx, 'MWh_buy'] = 200
    self.price.loc[idx, 'MWh_sell'] = 200
    return self.price

  def get_price(self, start_date, end_date, tariff = 'CAISO'):

    if(tariff == 'Transmission'):
      # Grid Opex
      # http://www.caiso.com/Documents/GridManagementChargeTraining.pdf
      # 0.0000836 -- Market Services Charge per KWh, both buyer and seller
      # 0.0002899 -- System Operations Charge per KWh, both buyer and seller
      # 0.005 -- Bid fee, excluded in v1

      self.price = self.get_caiso_price(start_date, end_date)

      return self.price

    elif (tariff == 'Distribution'):
      self.price = self.get_pge_b20_price(start_date, end_date)

      ## PDP Credits might be a nice option, but we are not eligible in the first year at least
      return self.price

    else:
      print("Unknown tariff")


  def get_co2_rate(self,start_date, end_date):
    df_co2_emissions = pd.DataFrame([])
    df_grid_demand = pd.DataFrame([])

    datetime_clc = start_date
    while datetime_clc < end_date:
        download_datetime = datetime_clc.strftime("%Y%m%d")
        datetime_clc += timedelta(days=1)
        print('now downloading: ' + download_datetime)

        url_co2 = 'http://www.caiso.com/outlook/SP/History/' + download_datetime + '/co2.csv'
        url_grid_demand = 'http://www.caiso.com/outlook/SP/History/' + download_datetime + '/demand.csv'

        df_co2_single = pd.read_csv(url_co2)
        periods_in_day = divmod(24 * 60, 5)[0]

        if (len(df_co2_single) < periods_in_day):
            missed_periods_num = periods_in_day - len(df_co2_single)
            print("missed periods:", missed_periods_num)
            df_co2_single = pd.concat([df_co2_single,
                                       pd.DataFrame(0, index=np.arange(missed_periods_num), columns=df_co2_single.columns)])

        df_co2_single['Time'].replace(['0:00'], '00:00', inplace=True)
        df_co2_emissions = pd.concat([df_co2_emissions, df_co2_single], ignore_index=True)

        df_grid_demand_single = pd.read_csv(url_grid_demand)
        df_grid_demand_single.drop(df_grid_demand_single.tail(1).index,inplace=True)
        df_grid_demand = pd.concat([df_grid_demand, df_grid_demand_single], ignore_index=True)

    df_co2_emissions = df_co2_emissions[df_co2_emissions['Time'] != df_co2_emissions['Time'].shift(1)]
    df_co2_emissions.to_csv(f'temp.csv', index=True)

    df_co2_emissions['TIME'] = pd.date_range(start_date, end_date, freq="5min", closed='left')
    df_co2_emissions.set_index('TIME', inplace=True)
    df_co2_emissions.drop('Time', axis=1, inplace=True)
    df_co2_emissions['CO2 rate'] = df_co2_emissions.sum(axis=1)
    df_co2_emissions = df_co2_emissions[['CO2 rate']]

    df_grid_demand['TIME'] = pd.date_range(start_date, end_date, freq="5min", closed='left')
    df_grid_demand.set_index('TIME', inplace=True)
    df_grid_demand = df_grid_demand.drop('Time', axis=1)
    df_grid_demand = df_grid_demand[['Current demand']]

    df_co2_emissions['CO2 rate'] = df_co2_emissions['CO2 rate'] / df_grid_demand['Current demand']
    return df_co2_emissions
```

```python
def get_electrolyzer_cost(electrolyzer_power, hydrogen_demand):
  df = pd.read_csv(root_path+'electrolyzer.csv', index_col=None, header=0, low_memory=False)

  cost = 0

  # hardware only cost
  capex = df.loc[(df["is_capex"] == "capex"), "Variable"].sum()
  #capex = df.loc[np.where((df["is_capex"] == "capex"))]["Variable"].sum()*electrolyzer_power

  # apply the tax shield
  #capex = capex*(1-electrolyzer["tax_shield"])
  #if (with_tax_shield == True):
  capex *= (1+economics["tax_state"])
  capex *= (1-electrolyzer["tax_shield"])

  capex *= electrolyzer_power

  #print("capex", capex)

  # installation part of the capex
  installation = df.loc[(df["is_capex"] == "installation", "Variable")].sum()
  installation *= electrolyzer_power
  installation *= (1+economics["tax_state"])

  #print("installation", installation)

  cost += (installation + capex)
  #print("capex+installation, el", cost)

  # electrolyzer change after 10 years
  # IRA Tax credit is not applicable in 10 years

  efficiency_parity = electrolyzer["FY22_efficiency"]/(electrolyzer["FY22_efficiency"] + electrolyzer["annual_efficiency_increase"]*electrolyzer["lifetime_years"])

  # 2nd purchase
  cost += ((capex+installation)*np.power((1-electrolyzer["annual_cost_drop"])/(1+economics["irr"]), electrolyzer["lifetime_years"])*efficiency_parity)

  # make scale adjustment
  if (electrolyzer_power > 20):
    cost *= np.power(0.989,electrolyzer_power/20)
  else:
    cost *= np.power(1.011,20/electrolyzer_power)

  # add opexes one by one, inflated under different rates
  opex = electrolyzer_power*df.loc[(df["is_capex"] == "opex"), "Variable"].sum()
  #print("opex", opex)
  opex *= ((1 - np.power(1/(1+economics["irr"]), project["lifetime"]+1))/(1 - 1/(1+economics["irr"])))
  opex *= (1+economics["tax_state"])

  # utility bills
  lifetime_multiplier = ((1 - np.power((1+economics["inflation"])/(1+economics["irr"]), project["lifetime"]+1))/(1 - (1+economics["inflation"])/(1+economics["irr"])))

  opex_kg = (hydrogen_demand*df.loc[(df["is_capex"] == "utilities"), "Per_KG"].sum())
  opex_kg *= (1+economics["tax_state"])

  cost += opex
  #print("system opex, el", opex)
  cost += opex_kg
  return cost
```

```python
def get_solar_cost(solar_power):
   df = pd.read_csv(root_path+'solar.csv', index_col=None, header=0, low_memory=False)

   # hardware only cost
   capex_fixed = df.loc[np.where((df["is_capex"] == "capex"))]["Fixed"].sum()
   capex_variable = df.loc[np.where((df["is_capex"] == "capex"))]["Variable"].sum()
   capex = capex_fixed+capex_variable*solar_power
   capex *= (1+economics["tax_state"])

   # annual costs
   opex_fixed = df.loc[np.where((df["is_capex"] == "opex"))]["Fixed"].sum()
   opex_variable = df.loc[np.where((df["is_capex"] == "opex"))]["Variable"].sum()
   opex = opex_fixed+opex_variable*solar_power

   # compute overall opex for the project lifetime
   opex *= (np.power((1-solar["annual_cost_drop"])/(1+economics["irr"]), project["lifetime"]+1)/(1 - (1-solar["annual_cost_drop"])/(1+economics["irr"])))
   opex *= (1+economics["tax_state"])

   # installation part of the capex
   installation_fixed = df.loc[np.where((df["is_capex"] == "installation"))]["Fixed"].sum()
   installation_variable = df.loc[np.where((df["is_capex"] == "installation"))]["Variable"].sum()
   installation = installation_fixed+installation_variable*solar_power
   installation *= (1+economics["tax_state"])

   capex_resale = solar["second_life_market_value"]*np.power((1-solar["annual_cost_drop"])/(1+economics["irr"]), project["lifetime"]+1)*capex

   capex *= (1-solar["tax_shield"])
   solar_cost = capex + installation + opex - capex_resale

   # compute LCOE
   lcoe = solar_cost/solar_power
   factor = (1 - np.power(1/(1 + economics["irr"]), project["lifetime"]+1))/(1-1/(1 + economics["irr"]))
   lcoe *= (1/(365*5.17*factor)) # median solar hours in this area, lifetime

   # solar_cost
   print("solar power, cost", solar_power, solar_cost)
   print("opex", opex)
   print("capex", capex)

   print("lcoe", lcoe)

   return solar_cost
```

```python
def get_battery_cost(battery_capacity):
  df = pd.read_csv(root_path+'battery.csv', index_col = None, header=0, low_memory = False)

  capex_unit = df["Capex_unit"].sum()
  opex_unit_year = df["Opex_unit"].sum()

  ## economy of scale

  # cost savings as a function of scale, 120 MWh
  scale_economy = 1
  if (battery_capacity < 120):
    scale_economy = (battery_capacity/120+((1-battery_capacity/120)*525/428))
  elif (battery_capacity < 240):
    scale_economy = ((battery_capacity-120)/120)*(369/428)+(240-battery_capacity)/120
  else:
    scale_economy = max(np.exp(5.97 - 0.00026*battery_capacity)/428, 1.1*215/428)

  # applied to both capex and opex
  capex  = capex_unit*scale_economy*battery_capacity

  ## assume the battery change once in 10 years, later on we fix it throught simulations with a "heavy technoeconomics"

  # get a cost of a new battery respecting the scale
  new_battery_cost = capex*np.power((1-battery["annual_cost_drop"])/(1+economics["irr"]), battery["lifetime"])

  # resell the old battery and add the new one to the capex
  capex += new_battery_cost
  capex *= (1+economics["tax_state"])

  capex_resale = new_battery_cost*battery["second_life_market_value"]

  # resell the battery at the end of the project
  # first resale
  capex_resale += np.power((1-battery["annual_cost_drop"])/(1+economics["irr"]), battery["lifetime"] + 1)
  capex_resale += np.power((1-battery["annual_cost_drop"])/(1+economics["irr"]), project["lifetime"] + 1)

  # account for the investment tax credit
  capex *= (1-battery["tax_shield"])

  # account for old battery reselling
  capex -= capex_resale

  #print("battery unit capex", capex/battery_capacity)

  # correction of scale and capacity
  opex = opex_unit_year*battery_capacity*scale_economy

  # correction for project lifetime
  opex *= (1 - np.power(1/(1+economics["irr"]), project["lifetime"]+1))
  opex *= (1/(1 - 1/(1+economics["irr"])))

  opex *= (1+economics["tax_state"])

  #print("battery unit opex", opex/battery_capacity)

  # we assume that the ITC correction is applied to Opex as well
  # as soon as (a) labor removed from the scope (b) maintenence can be treated as an upgrade and fall under the ITC scope
  #if (with_tax_shield):
  #  opex *= (1-battery["tax_shield"])

  battery_cost = capex + opex
  return battery_cost
```

```python
# under the assumption of a properly sized microgrid
def get_labor_cost(solar_power, electrolyzer_power):
  headcount = max(0.025*solar_power + 0.05*electrolyzer_power,3)
  annual_cost = headcount*economics["fte_cost"]

  # correct for the project timeline
  lifetime_correction = (1-np.power((1+economics["inflation"])/(1+economics["irr"]), project["lifetime"]+1))/(1 - (1+economics["inflation"])/(1+economics["irr"]))
  total_cost = annual_cost*lifetime_correction
  total_cost *= (1+economics["tax_state"])
  return total_cost
```

```python
# no energy cost is included
def get_connection_cost(pricing_plan, connection_power):
  df = pd.read_csv(root_path+'grid.csv', index_col=None, header=0, low_memory=False)
  capex_fixed = capex_variable = opex_fixed = opex_variable = 0

  # filter for the right pricing plan
  df = df[df['grid'] == pricing_plan]

  capex_fixed = df.loc[(df["is_capex"] == "capex") & (df["is_fixed"] == "fixed"), "cost"].sum()
  capex_variable = df.loc[(df["is_capex"] == "capex") & (df["is_fixed"] == "variable"), "cost"].sum()
  capex = capex_fixed + connection_power*capex_variable

  # apply a tax shield, a part of the opex may also be treated under the tax shield umbrella
  capex *= (1 - economics["tax_shield"])

  # securities is a part of capex, but the infrastructure credit is not applicable to them
  securities_fixed = df.loc[np.where((df["is_capex"] == "secirities") & (df["is_fixed"] == "fixed"))]["cost"].sum()
  securities_variable = df.loc[np.where((df["is_capex"] == "secirities") & (df["is_fixed"] == "variable"))]["cost"].sum()
  securities = securities_fixed + connection_power*securities_variable

  ## compute operational expenses

  # annual opex
  opex_fixed = df.loc[(df["is_capex"] == "opex") & (df["is_fixed"] == "fixed"), "cost"].sum()
  #df.loc[np.where((df["is_capex"] == "opex") & (df["is_fixed"] == "fixed"))]["cost"].sum()
  opex_variable = df.loc[(df["is_capex"] == "capex") & (df["is_fixed"] == "variable"), "cost"].sum()
  #[np.where((df["is_capex"] == "opex") & (df["is_fixed"] == "variable"))]["cost"].sum()
  opex = opex_fixed + opex_variable*connection_power

  #overall opex
  opex = opex*(1-np.power(1/(1+economics["irr"]), project["lifetime"]+1))/(1 - 1/(1+economics["irr"]))

  # real cost of security deposits
  securities_lifetime_corrector = 1 - np.power(1/(1+economics["irr"]), project["lifetime"])
  opex_lifetime_corrector = (1 - np.power(1/(1+economics["irr"]), project["lifetime"]))/(1 - 1/(1+economics["irr"]))

  if (pricing_plan == "caiso"):
    securities += securities_lifetime_corrector*(100000 if (connection_power < 20) else 250000)
  elif (pricing_plan == "b20"):
    # add the cost of transmission upgrade as needed
    capex += (0 if (connection_power < 20) else 7500000)
  elif (pricing_plan == "wdat"):
    securities += securities_lifetime_corrector*(100000 if (connection_power < 20) else 15000000)

  # the economy of scale -- potentially need multiple connections if too big
  connection_cost = securities + capex + opex

  if (pricing_plan == "b20"):
  # 110 MVA is a practical threshold, however scaling should be proportional to as a standard allowance 30 MVA
    connection_cost = connection_cost * max(connection_power/30.0, 1)
  else:
  # 750 MVA max is a practical threshold, however scaling should be proportional to as a standard allowance 45 MVA
    connection_cost = connection_cost * max(connection_power/45.0, 1)

  return connection_cost
```

```python
# create a default configuration to access the unit costs

def compute_unit_costs_volumes(airport, pricing_plan, h2_color):

  ##############################################################################
  ## Read the data from a file containing the grid price and the airport demand
  ##
  ## airport.csv -- contains "ORIGIN_AIRPORT", "GRID_COST", "SOLAR_HOURS" and
  ##                "KG_H2_DAY" columns
  ##
  ##############################################################################


  df = pd.read_csv(root_path+'airport.csv', index_col = None, header=0, low_memory=False)

  # filter by the airport we need
  df = df[df["ORIGIN_AIRPORT"] == airport]
  #print("df", df)

  # get solar hours
  solar_hours = df['SOLAR_HOURS'].values[0]
  #print("solar_hours", solar_hours)

  # get transmission grid cost =>> Change to demand_charge and energy_charge
  transmission_cost = df['TRANSMISSION_COST'].values[0]
  #print("transmission_cost", transmission_cost)

  # get overall demand
  h2_demand = df['KG_H2_DAY'].values[0]
  #print("h2_demand", h2_demand)


  ############################################################################
  ## Correction on the system degradation:
  ##
  ## Use maximal degradation among -- electrolyzer(2%), battery (2%), and
  ##                                  solar (0.5%)
  ##
  ############################################################################


  # system degradation stands for the maximal degradation over the components
  degradation_rate = 0.02 ## =>> change on the component wise
  degradation_correction = np.power(1-degradation_rate, project["renew_time"]+1)

  #print("degradation correction", degradation_correction)
  #print("h2 demand pre correction", h2_demand)

  # correct hydrogen request
  h2_demand *=  (1/degradation_correction)
  #print("h2 demand after correction", h2_demand)
  production["h2_demand"] = h2_demand


  # discounted volume of hydrogen production
  discounted_volume = economics["n_days"]*h2_demand
  discounted_volume *= (1 - np.power(1/(1+economics["irr"]), project["renew_time"]+1))/(1 - 1/(1+economics["irr"]))
  discounted_volume *= (project["lifetime"]/project["renew_time"])

  production["discounted_volume"] = discounted_volume

  # discounted volume
  #print("discounted_volume", discounted_volume)

  ############################################################################
  ## setup a default optimal configuration that later on will be improved
  ## The defaults are based on the a precomputed config with 10 tonn demsn
  ############################################################################

  # default demand to elсectrolyzer power standing for 24 hours of operations
  d2e_rate = 0.059

  # print
  print("electrolyzer default: ", h2_demand*d2e_rate/24)

  electrolyzer_power = h2_demand*d2e_rate/24
  solar_power = 4.8*electrolyzer_power #20*24/5
  battery_capacity = 24*electrolyzer_power


  ############################################################################
  ###### Plan and color specific optimization
  ############################################################################

  # levelized cost per unit of connection that is equal in power to electrolyzer
  uc = get_connection_cost(pricing_plan, electrolyzer_power)/(discounted_volume*electrolyzer_power)
  # levelized cost per unit of electrolyzer
  ue = get_electrolyzer_cost(electrolyzer_power, h2_demand)/(discounted_volume*electrolyzer_power)
  # levelized cost of solar per MW
  us = get_solar_cost(solar_power)/(discounted_volume*solar_power)
  #print("unit cost solar per MW", get_solar_cost(solar_power)/solar_power)
  # levelized cost of battery per MWh
  ub = get_battery_cost(battery_capacity)/(discounted_volume*battery_capacity)
  # levelized cost of grid per MWh of daily demand

  # labor cost. The labor cost is inflated with the IRS rate in time and discounted with the IRR
  ul = get_labor_cost(solar_power, electrolyzer_power)/discounted_volume
  #get_labor_cost(solar_power, electrolyzer_power)

  unit_costs["connection"] = uc
  default_configuration["connection"] = electrolyzer_power

  unit_costs["electrolyzer"] = ue
  default_configuration["electrolyzer"] = electrolyzer_power

  unit_costs["solar"] = us
  default_configuration["solar"] = solar_power

  unit_costs["battery"] = ub
  default_configuration["battery"] = battery_capacity

  unit_costs["labor"] = ul

```

```python
# The only reason against using PySAM Battery model is that
# I do not understand how to setaup trading having functionality they provide
#
# Recommend considering substitution this model to PySAM's one in future
#
class Battery:
  # All losses are roundtrip losses
  # All values in KW/KWh
  def __init__(self, connected_capacity, power_out_max = 0, power_in_max = 0,
               unit_loss = 0.024, inverter_loss = 0.03, capacity_at_replacement = 0.8,
               fce_current = 0, fce_max = 8000, om_year_rate = 0.015,
               overhead = 0.1, warranty_rate_year = 0.01,
               fce_day_max = 1, rate_inflation = RATE_INFLATION,
               battery_cost_drop_rate = 0.03, tax_credit_rate = 0.26):

    self.connected_capacity = connected_capacity

    ### Energy and losses

    # unit loss stands for loss with C = 1
    self.unit_loss = 0.024
    self.power_in_max = power_in_max

    self.power_in_max = power_in_max
    if (self.power_in_max == 0):
      # If the parameter is not specified, we assume the standard 4 hour battery
      self.power_in_max = self.connected_capacity/4

    self.power_out_max = power_out_max
    if (self.power_out_max == 0):
      # Inverter to storage ratio is setup to 1.67 according to https://atb.nrel.gov/electricity/2022/commercial_battery_storage#HCVLWBFD
      self.power_out_max = np.round(self.connected_capacity/1.67,2)

    self.C_factor = np.round(self.power_out_max/self.connected_capacity,2)

    # Used simplification, later to be changed w.r.t. to the outgoing power
    self.loss_roundtrip = unit_loss + inverter_loss

    # state of charge
    self.soc = 0.5
    self.soc_lower_bound = 0.1
    self.soc_upper_bound = 0.9


    ### Degradation

    #number of FCE's now
    self.fce_current = fce_current
    self.fce_max = fce_max
    self.capacity_at_replacement = capacity_at_replacement
    # introduce bound on the number of cycles
    self.fce_day_max = fce_day_max

    self.depreciation_year = ((350*fce_day_max)*(1 - capacity_at_replacement)/fce_max)
    self.depreciation_day = self.depreciation_year/350
    # TODO: correct it
    self.depreciation_cycle = self.depreciation_day


    ### Economics

    # Capex include installation and connection costs
    self.unit_capex = 0
    self.om_year_rate = om_year_rate
    # Warranty, overhead
    self.overhead = overhead
    self.warranty_rate_year = warranty_rate_year # annual

    # setup project lifetime
    self.lifetime = PROJECT_LIFETIME
    self.battery_cost_drop_rate = battery_cost_drop_rate

    self.get_capex(connected_capacity)
    #self.capex = self.get_capex(self.connnected_capacity)
    #self.om_year = om_year_rate*self.capex # annual

    # cost drop per year, better to substitute with NREL values
    self.battery_cost_drop_rate = battery_cost_drop_rate

    # Taxes

    self.tax_credit_rate = tax_credit_rate


  # NREL estimates have been used
  # https://atb.nrel.gov/electricity/2022/commercial_battery_storage
  # Should be elaborated in more details
  #
  # NREL provides estimates including labor etc. Does not match the estimates of TJ
  #
  def get_capex(self, connected_capacity):
    if (connected_capacity < 100):
      self.unit_capex = 2800
    elif(connected_capacity < 2000):
      self.unit_capex = 1200
    else:
      self.unit_capex = 900

    self.capex = self.unit_capex*self.connected_capacity

    # we do change the battery once in 7 years now
    # should be done based on actual degradation rate

    time_to_change = 7 #self.fce_max/(350*self.fce_day_max)
    i=1;

    while (i < (self.lifetime / time_to_change)):
      #print(i, self.lifetime / time_to_change)
      # price to buy a new battery
      self.capex = self.capex + (self.unit_capex*self.connected_capacity)*((1-self.battery_cost_drop_rate)**(i*time_to_change))

      # price to sell the old battery. 50% discount assumed
      self.capex = self.capex -(self.unit_capex*self.connected_capacity)*(((1 - self.depreciation_year)*(1-self.battery_cost_drop_rate))**(i*time_to_change))
      i = i+1

    # take annual charges into account
    # OM and Warranty
    # Inflation vs IRR
    # As investment,
    self.capex = self.capex + self.capex*(self.warranty_rate_year + self.om_year_rate)*(1-RATE_IRR**self.lifetime)/(1-RATE_IRR)

    # respect overhead
    self.capex = np.round(self.capex*(1+self.overhead),2)
    print("Battery MW vs capex", self.connected_capacity, self.capex)






```

```python
# TODO: IRR
class Solar:
  def __init__(self, connected_power, loss_roundtrip = 0.01, om_year_rate = 0.015,
               overhead = 0.1, warranty_rate_year = 0.01,
               rate_inflation = RATE_INFLATION,
               degradation_year = 0.005, solar_cost_drop_rate = 0.03):

    self.connected_power = connected_power
    self.solar_cost_drop_rate = solar_cost_drop_rate
    self.lifetime = PROJECT_LIFETIME
    ### Energy and losses

    # unit loss stands for loss with C = 1
    self.loss_roundtrip = loss_roundtrip
    self.overhead = overhead
    self.warranty_rate_year = warranty_rate_year
    self.om_year_rate = om_year_rate

    self.capex = 0
    self.get_capex()

  def get_capex(self):
    if (self.connected_power < 200):
      self.unit_capex = 2400
    elif(self.connected_power > 15000):
      self.unit_capex = 1000
    else:
      self.unit_capex = 1400

    self.capex = self.unit_capex*self.connected_power

    # we do change the battery once in 7 years now
    # should be done based on actual degradation rate

    time_to_change = 25 #self.fce_max/(350*self.fce_day_max)
    i=1;

    while (i < (self.lifetime / time_to_change)):
      #print(i, self.lifetime / time_to_change)
      # price to buy a new battery
      self.capex = self.capex + (self.unit_capex*self.connected_power)*((1-self.solar_cost_drop_rate)**(i*time_to_change))

      # price to sell the old battery. 50% discount assumed
      self.capex = self.capex - (self.unit_capex*self.connected_power)*(((1 - self.degradation_year)*(1-self.solar_cost_drop_rate))**(i*time_to_change))
      i = i+1

    # take annual charges into account
    # OM and Warranty
    # Inflation vs IRR
    # Discount to fulfil premium requirements on the rate of return
    self.capex = self.capex + self.capex*(self.warranty_rate_year + self.om_year_rate)*(1-RATE_IRR**self.lifetime)/(1-RATE_IRR)

    # respect overhead
    self.capex = np.round(self.capex*(1+self.overhead),2)
    print("Solar MW vs capex", self.connected_power, self.capex)
```

```python
# Pricing accoring to
# https://irena.org/-/media/Files/IRENA/Agency/Publication/2020/Dec/IRENA_Green_hydrogen_cost_2020.pdf
# Figure 26

class Electrolyser:
  def __init__(self, connected_power, loss_roundtrip = 0.02, om_year_rate = 0.015,
               overhead = 0.1, warranty_rate_year = 0.01,
               rate_inflation = RATE_INFLATION, installation_rate = 0.5,
               degradation_year = 0.023, lyser_cost_drop_rate = 0.05):

    self.connected_power = connected_power
    self.lyser_cost_drop_rate = lyser_cost_drop_rate
    self.lifetime = PROJECT_LIFETIME
    ### Energy and losses

    # unit loss stands for loss with C = 1
    self.loss_roundtrip = loss_roundtrip
    self.overhead = overhead
    self.warranty_rate_year = warranty_rate_year
    self.om_year_rate = om_year_rate
    self.installation_rate = installation_rate
    self.degradation_year = degradation_year
    self.min_power = connected_power/10

    self.capex = 0
    get_capex(self)
    self.KW_per_KG = 50

def get_capex(self):
    if (self.connected_power < 100):
      self.unit_capex = 1200
    elif(self.connected_power < 300):
      self.unit_capex = 900
    elif(self.connected_power < 300):
      self.unit_capex = 900
    elif(self.connected_power < 800):
      self.unit_capex = 800
    elif(self.connected_power < 1000):
      self.unit_capex = 750
    elif(self.connected_power < 2000):
      self.unit_capex = 650
    elif(self.connected_power < 3000):
      self.unit_capex = 600
    elif(self.connected_power < 5000):
      self.unit_capex = 550
    else:
      self.unit_capex = 500

    # 35 % for mounting, contingency, etc
    # 35 % for storage https://www.hydrogen.energy.gov/pdfs/review20/st100_houchins_2020_o.pdf
    self.capex = 1.7 * self.unit_capex*self.connected_power

    time_to_change = 10.5 #self.fce_max/(350*self.fce_day_max)
    i=1;
    while (i < (self.lifetime / time_to_change)):
      #print(i, self.lifetime / time_to_change)
      # price to buy a new battery
      self.capex = self.capex + (self.unit_capex*self.connected_power)*((1-self.lyser_cost_drop_rate)**(i*np.round(time_to_change,0)))

      # price to sell the old battery. 50% discount assumed
      self.capex = self.capex - (self.unit_capex*self.connected_power)*(((1 - self.degradation_year)*(1-self.lyser_cost_drop_rate))**(i*np.round(time_to_change,0)))
      i = i+1


    # Inflation vs IRR
    self.capex = self.capex + self.capex*(self.warranty_rate_year + self.om_year_rate)*(1-RATE_INFLATION**self.lifetime)/(1-RATE_INFLATION)

    # respect overhead
    self.capex = np.round(self.capex*(1+self.overhead),2)

```
